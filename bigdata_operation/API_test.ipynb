```python
url='https://app.rakuten.co.jp/services/api/IchibaItem/Search/20220601'
format='json'
applicationid='1018449994793832163'
keyword='うさまる'
url2=url+'?'+'format='+format+'&keyword='+urllib.parse.quote(keyword)+'&applicationId='+applicationid
print(url2)
print(url2=='https://app.rakuten.co.jp/services/api/IchibaItem/Search/20220601?format=json&keyword=%E6%A5%BD%E5%A4%A9&applicationId=1018449994793832163')
    
```

```python
import urllib.request as request
import json
response=request.urlopen(url2)
contents=response.read()
print(contents.decode())
```

```python
import pandas as pd
temp2=json.loads(contents.decode())

temp3=pd.json_normalize(temp2['Items'])# 阅读API的情报
temp3
```

```python
#读取所有页数的数据0
import time
usamaru_list=pd.DataFrame()# 导入一张空表
try:
    response=request.urlopen(url2) # 访问链接
    contents=response.read()  #读取链接 如果用requests函数只要requests.get就好 显示文字用text 显示图片用content
    response.close()
    temp=json.loads(contents.decode()) #decode 解码为json loads解码为可读
    temp10=pd.json_normalize(temp) #展开表格第一个参数是数据，可选record_path主要展开的路径，meta追加展开，没有选择则全部展开
    numofpages=temp10.pageCount
    for i in range(numofpages[0]):
        print(i+1,'page position')
        url3=url2+'&page'+str(i+1)
        time.sleep(0.5)
        response2=request.urlopen(url3)
        contents2=response2.read()
        tem4=json.loads(contents2.decode())
        temp5=pd.json_normalize(tem4['Items'])
        usamaru_list=pd.concat([usamaru_list,temp5],axis=0)
except urllib.error.HTTPError:
    print('not avaliable')
else:
    pass
usamaru_list

```

```python
shop_list=usamaru_list.groupby('Item.shopName')
shop_list.size()
```
